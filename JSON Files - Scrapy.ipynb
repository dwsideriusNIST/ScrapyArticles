```python
#Code to reformat the text in the JSON files 
#Specific to each publisher
#If run twice on same articles it will double many of the items 
import json
import re
import os
```

```python
def concatenate_list(input):
    output = ''
    for item in input:
        output = output+item
    return output
```

```python
pathacs = '/Users/juliachotoo/ScrapyArticles/scrapedjson/acs2/' #this will have to be changed to whatever folder the files are in

for filename in os.listdir(pathacs):
    with open(pathacs + filename, 'r'):
        resultacs = json.load(open(pathacs + filename, mode='r'))
        resultacs['text'] = concatenate_list(resultacs['text'])
        resultacs['text'] = resultacs['text'].split('ISSN', 2)
        resultacs['text'] = resultacs['text'][0]
        textacs = []
        textacs.append(resultacs['text'])
        resultacs['text'] = textacs
        for x in resultacs['text']:
            for y in resultacs['doi']:
                resultacs['text'] = y + " " + x
        resultacs['abstract'] = concatenate_list(resultacs['abstract'])
        resultacs['title'] = concatenate_list(resultacs['title'])
    with open(pathacs + filename, 'w') as jsonFile:
        json.dump(resultacs, jsonFile)
```

```python
#example of how corrected article should look
json.load(open(pathacs + '10.1021-Jp003626k.json', mode = 'r'))
```

```python
pathspr = '/Users/juliachotoo/ScrapyArticles/scrapedjson/spr2/' #this will have to be changed to whatever folder the files are in

for filename in os.listdir(pathspr):
    with open(pathspr + filename, 'r'):
        resultspr = json.load(open(pathspr + filename, mode='r'))
        resultspr['text'] = concatenate_list(resultspr['text'])
        text = []
        text.append(re.sub('\xa0', ' ', resultspr['text']))
        resultspr['text'] = text
        for x in resultspr['text']:
            for y in resultspr['doi']:
                resultspr['text'] = y + " " + x
        fixed_authors = []
        for x in resultspr['authors']:
            fixed_authors.append(re.sub('\xa0', ' ', x))
        resultspr['authors'] = fixed_authors
        resultspr['title'] = concatenate_list(resultspr['title'])
    with open(pathspr + filename, 'w') as jsonFile:
        json.dump(resultspr, jsonFile)
```

```python
#example of how the corrected article should look
json.load(open(pathspr + '10.1007-s10450-015-9695-3.json', mode = 'r'))
```
